# Operators

### Arithmatic Operators

```python
# Arithmatic Operators

print(f'Addition 11+23: {11+23}')
print(f'Subtraction 23-11: {23-11}')
print(f'Multiplication 11*23: {11*23}')
print(f'Division 23/11: {23/11}') # result with decimal part
print(f'Floor Division 23//11: {23//11}') # result without decimal part
print(f'Modulus 23%11: {23%11}') # remainder of the division
print(f'Exponentail 2^3: {2**3}')


```

### Comparators

```python
# Comparison Operators

# all produce boolean values (True or False)
print(f'Equal to 11==23: {11==23}')
print(f'Not Equal to 11!=23: {11!=23}')
print(f'Greater than 11>23: {11>23}')
print(f'Less than 11<23: {11<23}')
print(f'Greater than or equal to 11>=23: {11>=23}')
print(f'Less than or equal to 11<=23: {11<=23}')
print()

print(f'Length of "Python" and "Java": {len('Python') == len('Java')}')
print(f'Length of "mango" and "banana": {len('mango') > len('banana')}')
print(f'Length of "doll" and "ball": {len('doll') >= len('ball')}')
print()

# Additional Comparison Operators

print(f'Identity Operator: 3 is 3: {3 is 3}')
print(f'Identity Operator: 3 is not 4: {3 is not 4}')
print(f'Membership Operator: "a" in "banana": {"a" in "banana"}')
print(f'Membership Operator: "x" in "banana": {"x" in "banana"}')

```

### Logical Operators

```python
# Logical Operators

# AND operators
print(f'Logical AND [&&] - True and True: {True and True}')
print(f'Logical AND [&&] - True and False: {True and False}')
print(f'Logical AND [&&] - False and True: {False and True}')
print(f'Logical AND [&&] - False and False: {False and False}')
print()

print(f'3 > 2 and 5 > 4: {3 > 2 and 5 > 4}')
print(f'3 > 2 and 5 < 4: {3 > 2 and 5 < 4}')
print(f'3 < 2 and 5 > 4: {3 < 2 and 5 > 4}')
print(f'3 < 2 and 5 < 4: {3 < 2 and 5 < 4}')
print()

# OR operators
print(f'Logical OR [||] - True or True: {True or True}')
print(f'Logical OR [||] - True or False: {True or False}')
print(f'Logical OR [||] - False or True: {False or True}')
print(f'Logical OR [||] - False or False: {False or False}')
print()

print(f'3 > 2 or 5 > 4: {3 > 2 or 5 > 4}')
print(f'3 > 2 or 5 < 4: {3 > 2 or 5 < 4}')
print(f'3 < 2 or 5 > 4: {3 < 2 or 5 > 4}')
print(f'3 < 2 or 5 < 4: {3 < 2 or 5 < 4}')
print()

# NOT operators
print(f'Logical NOT [!]: not True: {not True}')
print(f'Logical NOT [!]: not False: {not False}')
print()

print(f'not (3 > 2): {not (3 > 2)}')
print(f'not (3 < 2): {not (3 < 2)}')
```
