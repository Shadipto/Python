```python
it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}
age = [22, 19, 24, 25, 26, 24, 25, 24] 

print(f'it_companies: {it_companies}')
print(f'A: {A}')
print(f'B: {B}')
print()

# 1.
print('1.')
joined_set = A.union(B)
print(f'Joined set A and B: {joined_set}')
print()

# 2.
print('2.')
print(f'Intersection of A and B: {A.intersection(B)}')
print()

# 3.
print('3.')
print(f'Is A subset of B?: {A.issubset(B)}')
print()

# 4.
print('4.')
print(f'Are A and B disjoint sets?: {A.isdisjoint(B)}')
print()

# 5.
print('5.')
print(f'Union of A with B: {A.union(B)}')
print(f'Union of B with A: {B.union(A)}')
print()

# 6.
print('6.')
print(f'Symmetric difference between A and B: {A.symmetric_difference(B)}')
print()

# 7.
print('7.')
print(f'Deleting the sets A and B')
del A
del B
try:
    print(A)
except NameError:
    print('A is deleted')
try:
    print(B)
except NameError:
    print('B is deleted')  
print()

```
