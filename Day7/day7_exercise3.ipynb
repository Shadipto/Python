### 3. I am a teacher and I love to inspire and teach people. How many unique words have been used in the sentence? Use the split methods and set to get the unique words.

```python
# sets
it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}
age = [22, 19, 24, 25, 26, 24, 25, 24]

print(f'it_companies: {it_companies}')
print(f'A: {A}')
print(f'B: {B}')
print(f'age: {age}')
print()

# 1.
print('1.')
print('Converting the age list to a set')
age_set = set(age)
print(f'age_set: {age_set}')
print(f'Comparing the length of age list and age_set: {len(age)} vs {len(age_set)}')
print()

# 2.

print('2.')
print('Difference data types of string, list, tuple and set:')
print('String: "Hello World"')
print('List: [1, 2, 3, 4, 5]')
print('Tuple: (1, 2, 3, 4, 5)')
print('Set: {1, 2, 3, 4, 5}')
print('String is a sequence of characters, list is an ordered collection of items, tuple is an ordered and immutable collection of items, and set is an unordered collection of unique items.')
print()

# 3.
print('3.')
string = 'I am a teacher and I love to inspire and teach people.'
print(f'Original string: {string}')
words_set = set(string.split(" "))
print(f'Number unique words in the string: {len(words_set)}')
print(f'Unique words in the string: {words_set}')
print()
```
