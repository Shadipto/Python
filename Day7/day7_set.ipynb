# Set

```python
# set [unique values in a list]
# declaed using set() function
# set are unordered, unindexed, and do not allow duplicate values

my_set = {1, 2, 3, 4, 5} # returns unique values in a list
print(f'my_set: {my_set}')
print(f'Data type of my_set: {type(my_set)}')
print()

# alternative declaration of a set using set() function

my_set2 = set([1, 2, 3, 4, 5]) # returns unique values in a list
print(f'my_set2: {my_set2}')
print()

print(f'Length of my_set: {len(my_set)}')
print(f'Length of my_set2: {len(my_set2)}')
print()

# frozenset is an immutable set, which means that it cannot be modified after it is created

my_frozenset = frozenset([1, 2, 3, 4, 5]) # returns unique values in a list
print(f'my_frozenset: {my_frozenset}')
print(f'Data type of my_frozenset: {type(my_frozenset)}')
print()

# mixed data types in a set

mixed = {1, 'Hello', 3.14, (1, 2, 3), frozenset({4, 5, 6})}
print(f'mixed: {mixed}')
print(f'Data type of mixed: {type(mixed)}')
print(f'Data type of index 0 of mixed: {type(frozenset({4, 5, 6}))}')
print()
```

## Acccessing set elements

```python
# accessing set items
# set items cannot be accessed by index, but we can loop through the set items using a

my_set = {1, 2, 3, 4, 5}
print('Looping through my_set:')
for item in my_set:
    print(item)
print()

# checking if an item is in a set using the 'in' keyword
print(f'Is 3 in my_set? {3 in my_set}')
print(f'Is 6 in my_set? {6 in my_set}')
print()

# adding items to a set using the add() method
my_set.add(6)
print(f'my_set after adding 6: {my_set}') # cant use append() method to add items to a set
print()

# adding multiple items to a set using the update() method
my_set.update([7, 8, 9])
print(f'my_set after adding multiple items: {my_set}')
print()

# removing items from a set using the remove() method
my_set.remove(9)
print(f'my_set after removing 9: {my_set}')
print()
my_set.remove(2)
print(f'my_set after removing 2: {my_set}')

# pop() method removes a random item from the set and returns it
my_set.pop() # removes a random item from the set and returns it
print(f'Popped item form my_set: {my_set}')
print(f'my_set after popping an item: {my_set}')
print()

# clear() method removes all items from the set

print(f'my_set before clearing: {my_set}')
my_set.clear() # removes all items from the set
print(f'my_set after clearing: {my_set}')
print()

# deleting a set using the del keyword
my_set2 = {1, 2, 3, 4, 5}
print(f'my_set2 before deleting: {my_set2}')
del my_set2 # deletes the set
try:
    print(f'my_set2 after deleting: {my_set2}') # raises an error because my_set2 is deleted
except NameError:
    print('my_set2 is deleted and cannot be accessed')
```

## Converting set to list/tuple

```python
# set does not allow duplicate values
my_set = {1, 2, 3, 4, 5, 5, 5, 5} # returns unique values in a list
print(f'my_set with duplicate values: {my_set}') # returns unique values in a list
print()

my_list = my_set # converting a set to a list
print(f'my_list: {my_list}')
print(f'Data type of my_list: {type(my_list)}')
print()
```

## Functions of set

```python
# joining sets using the union() method

set1 = {1, 2, 3}
set2 = {3, 4, 5}
set3 = set1.union(set2) # returns a new set that contains all the items from both sets, without duplicates
print(f'set1: {set1}')
print(f'set2: {set2}')
print(f'set3 (union of set1 and set2): {set3}')
print()

# intersection of sets using the intersection() method
set4 = set1.intersection(set2) # returns a new set that contains only the items that are present in both sets
print(f'set4 (intersection of set1 and set2): {set4}')
print()

# subset and superset

set5 = {1, 2}
print(f'set5: {set5}')
print(f'Is set5 a subset of set1? {set5.issubset(set1)}') # returns True if set5 is a subset of set1
print(f'Is set1 a superset of set5? {set1.issuperset(set5)}') # returns True if set1 is a superset of set5
print()

# difference of sets using the difference() method
set6 = set1.difference(set2) # returns a new set that contains only the items that are present in set1 but not in set2
print(f'set6 (difference of set1 and set2): {set6}')
print()

# disjoint sets using the isdisjoint() method
set7 = {6, 7, 8}
print(f'set7: {set7}')
print(f'Is set1 disjoint with set7? {set1.isdisjoint(set7)}') # returns True if set1 and set7 have no items in common
print()
```
