### it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
### A = {19, 22, 24, 20, 25, 26}
### B = {19, 22, 20, 25, 26, 24, 28, 27}
### age = [22, 19, 24, 25, 26, 24, 25, 24]

### 1. Find the length of the set it_companies

### 2. Add 'Twitter' to it_companies

### 3. Insert multiple IT companies at once to the set it_companies

### 4. Remove one of the companies from the set it_companies

### 5. What is the difference between remove and discard

```python
it_companies = {'Facebook', 'Google', 'Microsoft', 'Apple', 'IBM', 'Oracle', 'Amazon'}
A = {19, 22, 24, 20, 25, 26}
B = {19, 22, 20, 25, 26, 24, 28, 27}
age = [22, 19, 24, 25, 26, 24, 25, 24]

print(f'it_companies: {it_companies}')
print(f'A: {A}')
print(f'B: {B}')
print(f'age: {age}')
print()

# 1.

print('1.')
print(f'Length of it_companies: {len(it_companies)}')
print()

# 2.

print('2.')
it_companies.add('Twitter')
print(f'it_companies after adding Twitter: {it_companies}')
print()

# 3.

print('3.')
it_companies.update(['Meta', 'LinkedIn', 'Netflix'])
print(f'it_companies after adding Meta, LinkedIn, Netflix: {it_companies}')
print()

# 4.

print('4.')
it_companies.pop()
print(f'it_companies after removing the last company: {it_companies}')
print()

# 5.

print('5.')
print(f'Difference between remove and discard: remove() will raise an error if the item is not found, while discard() will not raise an error if the item is not found.')
print()
```
