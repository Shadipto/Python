## 2. Conduct the following operations. The operands are 3 and 4.
### addition(+)
### subtraction(-)
### multiplication(*)
### modulus(%)
### division(/)
### exponential(**)
### floor division operator(//)

```python
print(f"addition of 3 and 4: {3 + 4}" )

print(f"subtraction of 3 from 4: {4 - 3}")

print(f"multiplication of 3 and 4: {3 * 4}")

print(f"modulus of 3 and 4: {3 % 4}")

print(f"division of 3 by 4: {3 / 4}")

print(f'exponential of 3 to the power of 4: {3 ** 4}')

print(f'floor dividion of 3 by 4: {3 // 4}')
```

# Find an Euclidean distance between (2, 3) and (10, 8)

```python
# (2,3)

x1 = 2
y1 = 3

# (10,8)

x2 = 10 
y2 = 8

# euclidean distance = sqrt((x2 - x1)^2 + (y2 - y1)^2)

import math

print(f"Euclidean distance between point1({x1},{y1}) and point2({x2},{y2}) is: {math.sqrt((x2 - x1)**2 + (y2 - y1)**2)}")
```
