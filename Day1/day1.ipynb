# Datatypes in Python

## Integer

```python
#Integer

a = 10
b = 6
c = 22

print(a)
print(b)
print(c)
```

## Float

```python
#Float

d = 2.4
e = 34.23
pi = 3.1416

print(d)
print(e)
print(pi)
```

## String

```python
#String

f = "hello" # string can be defined by single or double quote
g = 'world'
h = "hudai likhlam"

print(f)
print(g)
print(h)
```

## Boolean

```python
# Boolean

aa = True
bb = False

print(aa)  
print(bb)
```

# Complex

```python
# Complex

x = 4 + 3j
y = 5j

print(x)
print(y)
print(type(x))
print(type(y))
print(x.real)
print(x.imag)
```

## List

```python
# List

my_list = [1,2,3,4,5,6] # [] is used to define list

my_list2 = ["apple", "banana", "kacha mango", 'papaya'] # list can contain strings

my_list3 = [1, 3.1416, True, "apple", 'bus'] # list can contain different data types

print(my_list)
print(my_list2)
print(my_list3)

print(my_list[0]) # list index starts from 0 upto n-1
print(my_list[1]) 
print(my_list[2])
```

## Dictionary

```python
# Dictionary

### Declared by {}
# It's a collection of key-value pairs
# It is unordered and mutable (can be changed)
# It can hold different data types [even other dictionaries and lists]
# Can have duplicate values but not duplicate keys

x = {
    "name": 'Dipto',
    "age": 22,
    "is_student": True,
    'hobbies': ['coding', 'gaming', 'traveling'] ## can use "" or '' for keys and values
}

print(x)
print(x["name"])
print(x["age"])
print(x["is_student"])
print(x["hobbies"])
print(x["hobbies"][0]) # accessing first index of hobbies list
print(x["hobbies"][1]) # accessing second index of hobbies list
print(x["hobbies"][2]) # accessing third index of hobbies list

x['name'] = "new name" # changing the value of name key
print(x['name'])
x['hobbies'].append('new hobby poetry') # adding new hobby to hobbies list
print(x['hobbies'])
```

## Tuple

```python
# Tuple

### Declared by ()
# It's an ordered collecton of items
# It cannot be changed (immutable)
# It can hold different data types [even other tuples and lists]
# Can have duplicate values

x = (1, 11, 11, 22, 45, 23)

y = ("apple", "banana", 'kacha mango', 'papaya')

z = (1, 3.1416, True, "apple", 'bus')

print(x)
print(x[0]) # tuple index starts from 0 upto n-1
print(x[1]) 
print(y)
print(z)
```

## Set

```python
# Set

### Declared by {}
# It's an unordered collection of unique items
# It is mutable (can be changed)
# It can hold different data types [even other sets and lists]

my_set ={1, 2, 3, 4, 5}

my_set2 = {"apple", "banana", 'kacha mango', 'papaya'}

my_set3 = {1, 3.1416, True, "apple", 'bus'}

print(my_set)
print(my_set2)
print(my_set3)

# set does not support indexing
# print(my_set[0]) ==> this will give error because set does not support indexing
```

## Type checking
