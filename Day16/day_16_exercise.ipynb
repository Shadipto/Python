# Exercise

### 1. Get the current day, month, year, hour, minute and timestamp from datetime module

### 2. Format the current date using this format: "%m/%d/%Y, %H:%M:%S"

### 3. Today is 5 December, 2019. Change this time string to time.

### 4. Calculate the time difference between now and new year.

### 5. Calculate the time difference between 1 January 1970 and now.


```python
# 1
from datetime import datetime
print('1.')
current_datetime = datetime.now()

print(f'Current date and time: {current_datetime}')
print(f'Current day: {current_datetime.day}')
print(f'Current month: {current_datetime.month}')
print(f'Current hour: {current_datetime.hour}')
print(f'Current minute: {current_datetime.minute}')

timeStamp = current_datetime.timestamp()
print(f'Timestamp: {timeStamp}\n')

# 2
print('2.')

print(f'Formatted in %m/%d/%Y, %H:%M:%S: {current_datetime.strftime("%m/%d/%Y, %H:%M:%S")}\n')

# 3
print('3.')
string = "Today is 5 December, 2019"

date_obj = datetime.strptime(string, 'Today is %d %B, %Y') # converts the string to a datetime object using the specified format %b for month name, %d for day, and %Y for year
print(f'Date object: {date_obj}\n')

# 4.
print('4.')
new_year = datetime(2027, 1, 1) # creates a datetime object for January 1, 2027
print(f'Time until New Year: {new_year - current_datetime}')
print(f'Time until New Year in days: {(new_year - current_datetime).days}\n')

# 5.
print('5.')
old_date = datetime(1970, 1, 1) 
print(f'Time since January 1, 1970: {(current_datetime - old_date).days} days\n')
```
