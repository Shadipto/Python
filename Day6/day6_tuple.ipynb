# Tuple

```python
# Tuple (immutable list)
# Tuples are defined using () instead of []

my_tuple = (1, 2, 3, 4, 5)

print(f'My tuple: {my_tuple}')
print()

# Accessing elements in a tuple
print(f'First element: {my_tuple[0]}')
print(f'Last element: {my_tuple[-1]}') # using negative indexing
print()

# empty tuple

empty_tuple = ()
print(f'Empty tuple: {empty_tuple}')
print()

mixed = ('hello', 42, 3.14, [1, 2, 3], (4, 5))
print(f'New tuple with mixed data types: {mixed}')
print(f'Length of mixed tuple: {len(mixed)}')
print()

```

## Slicing tuple

```python
# slicing a tuple
test = ('saturday', 'sunday', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday')
print(f'Test tuple: {test}')
print(f'Slice from index 2 to 5: {test[2:5]}')
print(f'Slice from index 0 to 3: {test[:3]}')
print(f'Slice from index 4 to the end: {test[4:]}')
print(f'Slicing the last 3 elements: {test[-3:]}')
print()
```

## Converting tuple to list

```python
# converting tuple to list and vice versa

tupleA = (1, 2, 3, 4, 5)
listA = list(tupleA) # converting tuple to list

print(f'List converted from tupleA to listA: {listA}') # now we can modify listA
listA.append(6) # adding an element to the list
print(f'ListA after appending 6: {listA}')
print(f'TupleA remains unchanged: {tupleA}') # tupleA is still the same
print()
```

## Joining tuple

```python
# joining two tuples
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)
joined_tuple = tuple1 + tuple2

print(f'Tuple1: {tuple1}')
print(f'Tuple2: {tuple2}')
print(f'Joined tuple: {joined_tuple}')
print()
```
