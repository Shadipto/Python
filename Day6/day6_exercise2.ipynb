# Exercise Level2

### 1. Unpack siblings and parents from family_members

### 2. Create fruits, vegetables and animal products tuples. Join the three tuples and assign it to a variable called food_stuff_tp.

### 3. Change the about food_stuff_tp tuple to a food_stuff_lt list

### 4. Slice out the middle item or items from the food_stuff_tp tuple or food_stuff_lt list.

### 5. Slice out the first three items and the last three items from food_stuff_lt list

### 6. Delete the food_stuff_tp tuple completely

```python
family = ('Eliza', 'Sophia', 'Mia', 'Liam', 'Noah', 'Ethan', 'Jane', 'John')

# 1.

print('1.')
print(f'Family members: {family}')
print(f'Unpacking siblings: {family[:6]}')
print(f'Unpacking parents: {family[6:]}')
print()

# 2.

print('2.')
fruits = ('apple', 'banana', 'cherry', 'date', 'elderberry')
print(f'Fruits: {fruits}')
vegetables = ('carrot', 'broccoli', 'spinach', 'pepper', 'cucumber')
print(f'Vegetables: {vegetables}')
animal_products = ('milk', 'cheese', 'yogurt', 'butter', 'cream')
print(f'Animal products: {animal_products}')
print()

food_stuff = fruits + vegetables + animal_products
print(f'Food stuff (joining all): {food_stuff}')
print()

# 3.

print('3.')
print(f'Type of food_stuff: {type(food_stuff)}')
food_stuff_lt = list(food_stuff)
print(f'Food stuff converted to list: {food_stuff_lt}')
print(f'Type of food_stuff_lt: {type(food_stuff_lt)}')
print()

# 4.

print('4.')
print(f'Slicing the middle item from the tuple: {food_stuff[len(food_stuff) // 2]}')
print()

# 5.

print('5.')
print(f'First three items from the list: {food_stuff_lt[:3]}')
print(f'Last three items from the list: {food_stuff_lt[-3:]}')
print()

# 6.

print('6.')
del food_stuff
try:
    print(food_stuff)
except NameError:
    print('food_stuff has been deleted and is no longer accessible.')

print()
```
