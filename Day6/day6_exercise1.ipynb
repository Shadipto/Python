# Exercise Level1

### 1. Create an empty tuple

### 2. Create a tuple containing names of your sisters and your brothers (imaginary siblings are fine)

### 3. Join brothers and sisters tuples and assign it to siblings

### 4. How many siblings do you have?

### 5. Modify the siblings tuple and add the name of your father and mother and assign it to family_members

```python
# 1.

empty_tuple = ()
print('1.')
print(f'Empty tuple: {empty_tuple}')
print()

# 2.

print('2.')
sister_tuple = ('Eliza', 'Sophia', 'Mia')
brother_tuple = ('Liam', 'Noah', 'Ethan')
print(f'Siblings tuple: {sister_tuple}')
print(f'Brothers tuple: {brother_tuple}')
print()

# 3.

print('3.')
siblings_tuple = sister_tuple + brother_tuple
print(f'All siblings tuple: {siblings_tuple}')
print()

# 4.

print('4.')
print(f'Total number of siblings: {len(siblings_tuple)}')
print()

# 5.

print('5.')
mother = 'Jane'
father = 'John'
family_members = siblings_tuple + (mother, father) # cant do siblings_tuple + mother + father because they are not tuples, we need to convert them to tuples first
print(f'Family members tuple: {family_members}')
print(f'Total number of family members: {len(family_members)}')
print()
```
