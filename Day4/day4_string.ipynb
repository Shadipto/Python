# Strings

```python
# Strings

string = 'A' # '' or "" can be used to define a string
print(f'string: {string}')
print(type(string))
print(f'Length of string: {len(string)}')
print()

greeting_string = "Hello, World!"
print(f'greeting_string: {greeting_string}')
print(type(greeting_string))
print(f'Length of greeting_string: {len(greeting_string)}')
print()

greeting_string2 = 'How are you?'
print(f'greeting_string2: {greeting_string2}')
print(type(greeting_string2))
print(f'Length of greeting_string2: {len(greeting_string2)}')
print()

```

## Multiline Strings

```python
# Multiline string

# Multiline string can be defined using triple quotes (''' or """)
multiline_string = '''Hello, world!
Ki obostha?
Ami bhalo nei.'''

print(f'''multiline_string: 
{multiline_string}''')

print(type(multiline_string))
print(f'Length of multiline_string: {len(multiline_string)}')
print()
```

## String Contatination

```python
# String concatenation

firstName = 'Shadipto'
lastName = "Pranto"
space = ' '

fullName = firstName + space + lastName
print(f'Full Name: {fullName}')
print(type(fullName))
print(f'Length of fullName: {len(fullName)}')
print()

print(f'Length of firstName: {len(firstName)}')
print(f'Length of lastName: {len(lastName)}')
print(f'Length of space: {len(space)}')
print(f'Length of fullName: {len(fullName)}') # space will be counted as well
print()
```

## String escape sequences

```python
# string escape sequences

# newline [\n]
print('Hello, world! \nKi obostha? \nAmi bhalo nei. ') # everything after \n will be printed in a new line [even spaces]
print()

# tab [\t] 4 spaces
print('Hello, world! \tKi obostha? \tAmi bhalo nei. ') # everything after \t [4 spaces] will be printed in a new line [even spaces]
print('Date:\t07\tFebruary\t2026')
print('Shadipto\tPranto')
print()

# backslash [\\] to print a backslash(\), we need to use double backslash [\\]
print('This is a backslash: \\')
date = 'Date: 07\\Feb\\2026'
print(f"Date: {date}")
print()

# Double quotes inside a string
print("She said, \"Hello, World!\"") # to print double quotes inside a string, we need to use backslash [\] before the double quotes
print('She said, "Hello, World!"') # we can also use single quotes to define the string, then we don't need to use backslash [\] before the double quotes
print()
```

## Old string formatting

```python
# old string formatting

# %s is used for string formatting [any object with a string representation, like numbers]
name = 'Shadipto'
date = '07 February 2026'
print('Hello, my name is %s.' %(name)) # %s will be replaced by the value of name
print('My name is %s and today is %s.' %(name, date)) # %s will be replaced by the value of name and date respectively
print()

# %d is used for integer formatting [object with an integer representation, like integer(not float)]
age = 22
print('My name is %s and I am %d years old.' %(name, age)) # %d will be replaced by the value of age
print()

# %f is used for float formatting [float objects and any object with a float representation, like float]
pi = 3.14159
print('The value of pi is approximately %f.' %(pi)) # %f will be replaced by the value of pi
print('The value of pi is approximately %.2f.' %(pi)) # %.2f will round the value of pi to 2 decimal places
print()
```

## Style string formatting

```python
# style string formatting

# {} is used for string formatting
name = 'Shadipto'
date = '07 February 2026'
print('Hello, my name is {}.'.format(name)) # {} will be replaced by the value of name
print('My name is {} and today is {}.'.format(name, date)) # {} will be replaced by the value of name and date respectively
print()

age = 22
print('My name is {} and I am {} years old.'.format(name, age)) # {} will be replaced by the value of name and age respectively
print()

pi = 3.14159
print('The value of pi is approximately {}.'.format(pi)) # {} will be replaced by the value of pi
print('The value of pi is approximately {:.2f}.'.format(pi)) # {:.2f} will round the value of pi to 2 decimal places
print()

import math
radius = 5
print('Radius of the circle: {}.' .format(radius))
print('Area of the circle: {:.2f}.' .format(pi * math.pow(radius, 2))) 
print()
```

# f-string formatting

## Indexing strings

```python
# indexing strings

testing = 'Python'

print(testing[0]) # P
print(testing[1]) # y
print(testing[2]) # t
print(testing[3]) # h
print(testing[4]) # o
print(testing[5]) # n
print()

print(f'Last index of "Python": {len(testing) - 1}') # first index is 0 always
print(f'Number of letters in "Python": {len(testing)}')
print()

# negative indexing

# starting is from -1 not 0
print(testing[-1]) # n
print(testing[-2]) # o
print(testing[-3]) # h
print(testing[-4]) # t
print(testing[-5]) # y
print(testing[-6]) # P
print()

print(f'Second last letter of "Python": {testing[-2]}')
```

## Slicing python strings

```python
# slicing strings

string = 'Python Programming'

print(string)
# [start_index:final_index] meaning start index to final index (excluding final index)

print(f'First 5 letters: {string[0:5]}') # first 5 letters will be printed [0 to 4]
print(f'Last 4 letters: {string[-4:]}') # last 4 letters will be printed [-4 to -1], dont do string[-4:-1 or 0](wrong syntax)
# another way 
print(f'First 5 letters: {string[:5]}') 

```

## Reversing string

```python
# reversing a string

string = 'Python Programming'

print(f'Original string: {string}')
print(f'Reversed string: {string[::-1]}') # [start:final:step] step is -1 means reverse the string
print()

# skipping characters in a string

print(f'Original string: {string}')
print(f'Skipping alternate characters: {string[::2]}') # [start:final:step] step is 2 means skip alternate characters
print()
```

## String Methods

### capitalization()

```python
# string methods

# capitalize() proper form
string = 'python programming'

print(f'Original string: {string}')
print(f'Capitalized string: {string.capitalize()}') # capitalize() will capitalize the first letter of the string
print()
```

### count()

```python
# count() counts the number of occurrences of a substring in a string
print(f'Original string: {string}')
print(f'Number of "o" in the string: {string.count('o')}') # count() will count the number of occurrences of "o" in the string
print(f'Number of "th" in the string: {string.count('th')}') 
print(f'Number of "z" in the string: {string.count('z')}') 
print(f'Number of \' \' in the string: {string.count(' ')}') # used \' \' to print single quotes inside a string

# count() will count the number of occurrences of "g" in the string from index 7 to 19
print(f'Number of \'g\' in the string from index 7 to 19: {string.count('g', 7, 19)}') 
print(f'Number of \'m\' in the string from index 7 to 8: {string.count('g', 7, 16)}') # included index 7 but did not included index 16
print()
```

### endswith()

```python
# endswith() checks if the string ends with a specified suffix

print(f'Original string: {string}')
print(f'Does the string end with "ing"?: {string.endswith('ing')}')
print(f'Does the string end with "Python"?: {string.endswith('Python')}')
print()
```

### expandtabs()

```python
# expandtabs() expands the tabs in a string to spaces base tab size = 4

string = 'Hello friends.\tHow are you?\tI hope you are doing well.' # must use \t to print tab in the string

print(f'Original string: {string}')
print()
print(f'String with expanded tabs: {string.expandtabs(10)}') # expandtabs() will expand the tabs in the string to spaces base tab size = 10 (it can be any number)
print()
```

### find()

```python
# find() returns the lowest index of the substring if it is found in the string. If it is not found, it returns -1

string = 'Python Programming'
print(f'Original string: {string}')
print(f'Index of "m": {string.find('m')}')
print(f'Index of " " (space): {string.find(' ')}')
print(f'Index of "z": {string.find('z')}') # "z" is not found in the string, so it returns -1
print(f'Index of "mm": {string.find('mm')}') # "mm" is found in the string, so it returns the index of the first "m"
print()
```

### rfind()

```python
# rfind() returns the last index of the substring if it is found in the string. If it is not found, it returns -1

string = 'Python Programming'
print(f'Original string: {string}')
print(f'Last index of "m": {string.rfind('m')}') # "m" is found in the string, so it returns the index of the last "m"
print(f'Last index of " " (space): {string.rfind(' ')}')
print(f'Last index of "z": {string.rfind('z')}') # "z" is not found in the string, so it returns -1
print()
```

### index()

```python
# index() is similar to find() but it raises a ValueError if the substring is not found in the string 

string = 'Python Programming'
print(f'Original string: {string}')
print(f'Index of "m": {string.index('m')}')
print(f'Index of " " (space): {string.index(' ')}')
print(f'Index of "P": {string.index('P')}') # here 'P' is case sensitive, so it will return the index of 'P' not 'p'
#print(f'Index of "p": {string.index('p')}') # here 'p' is case sensitive this will return error
print()
```

### isalnum()

```python
# isalnum() checks if all the characters in the string are alphanumeric (letters and numbers only) [a-z, A-Z, 0-9]
# no spaces or special characters allowed

string1 = 'Python123'
string2 = 'Python 123'
string3 = 'Python@123'

print(f'Is "{string1}" alphanumeric?: {string1.isalnum()}') # all characters in string1 are alphanumeric, so it returns True
print(f'Is "{string2}" alphanumeric?: {string2.isalnum()}') # string2 contains a space, so it returns False
print(f'Is "{string3}" alphanumeric?: {string3.isalnum()}') # string3 contains a special character (@), so it returns False
print()
```

### isalpha()

```python
# isalpha() checks if all the characters in the string are alphabetic (letters only) [a-z and A-Z]
# no spaces, numbers or special characters allowed

string1 = 'Python'
string2 = 'Python 123'
string3 = 'Python@123'

print(f'Is "{string1}" alphabetic?: {string1.isalpha()}') # all characters in string1 are alphabetic, so it returns True
print(f'Is "{string2}" alphabetic?: {string2.isalpha()}') # string2 contains a space and numbers, so it returns False
print(f'Is "{string3}" alphabetic?: {string3.isalpha()}') # string3 contains a special character (@) and numbers, so it returns False
print()
```

### isdicimal()

```python
# isdecimal() checks if all the characters in the string are decimal characters (numbers only) [0-9]
# no spaces, letters or special characters allowed

string1 = '123456'
string2 = '123 456'
string3 = '123@456'

print(f'Is "{string1}" decimal?: {string1.isdecimal()}') # all characters in string1 are decimal, so it returns True
print(f'Is "{string2}" decimal?: {string2.isdecimal()}') # string2 contains a space, so it returns False
print(f'Is "{string3}" decimal?: {string3.isdecimal()}') # string3 contains a special character (@), so it returns False
print()
```

### isdigit()

```python
# isdigit() checks if all the characters in the string are digits (numbers only) [0-9 and some other characters like superscript and subscript numbers]

string1 = '123456'
string2 = '123 456'
string3 = '123@456'
string4 = '²³' # superscript numbers
string5 = '₁₂' # subscript numbers

print(f'Is "{string1}" digit?: {string1.isdigit()}') # all characters in string1 are digits, so it returns True
print(f'Is "{string2}" digit?: {string2.isdigit()}') # string2 contains a space, so it returns False
print(f'Is "{string3}" digit?: {string3.isdigit()}') # string3 contains a special character (@), so it returns False
print(f'Is "{string4}" digit?: {string4.isdigit()}') # all characters in string4 are digits (superscript numbers), so it returns True
print(f'Is "{string5}" digit?: {string5.isdigit()}') # all characters in string5 are digits (subscript numbers), so it returns True
print()
```

### isnumeric()

```python
# isnumeric() checks if all the characters in the string are numeric characters (numbers only) [0-9 and some other characters like superscript and subscript numbers]

string1 = '123456'
string2 = '123 456'
string3 = '123@456'
string4 = '²³' # superscript numbers
string5 = '₁₂' # subscript numbers
string6 = '⅓' # fraction
string7 = '½' # fraction


print(f'Is "{string1}" numeric?: {string1.isnumeric()}') # all characters in string1 are numeric, so it returns True
print(f'Is "{string2}" numeric?: {string2.isnumeric()}') # string2 contains a space, so it returns False
print(f'Is "{string3}" numeric?: {string3.isnumeric()}') # string3 contains a special character (@), so it returns False
print(f'Is "{string4}" numeric?: {string4.isnumeric()}') # all characters in string4 are numeric (superscript numbers), so it returns True
print(f'Is "{string5}" numeric?: {string5.isnumeric()}') # all characters in string5 are numeric (subscript numbers), so it returns True
print(f'Is "{string6}" numeric?: {string6.isnumeric()}') # all characters in string6 are numeric (fraction), so it returns True
print(f'Is "{string7}" numeric?: {string7.isnumeric()}') # all characters in string7 are numeric (fraction), so it returns True
print()
```

### isidentifier()

```python
# isidentifier() checks if the string is a valid identifier (variable name) or not
# a valid identifier must start with a letter (a-z or A-Z) or an underscore (_), followed by letters, digits (0-9) or underscores (_)
# a valid identifier cannot be a reserved keyword in Python (like if, else, for, while, etc.)

string1 = 'variable_name'
string2 = 'variable name'
string3 = '1variable_name'
string4 = '_variable_name'
string5 = 'if' # reserved keyword
print(f'Is "{string1}" a valid identifier?: {string1.isidentifier()}') # string1 is a valid identifier, so it returns True
print(f'Is "{string2}" a valid identifier?: {string2.isidentifier()}') # string2 contains a space, so it returns False
print(f'Is "{string3}" a valid identifier?: {string3.isidentifier()}') # string3 starts with a digit, so it returns False
print(f'Is "{string4}" a valid identifier?: {string4.isidentifier()}') # string4 starts with an underscore, so it returns True
print(f'Is "{string5}" a valid identifier?: {string5.isidentifier()}') # string5 is a reserved keyword, so it returns False
print()
```

### islower()

```python
# islower() checks if all the characters in the string are lowercase letters (a-z) and there is at least one character in the string

string1 = 'python'
string2 = 'Python'
string3 = 'python123'
string4 = 'python@123'
string5 = '' # empty string

print(f'Is "{string1}" lowercase?: {string1.islower()}') # all characters in string1 are lowercase letters, so it returns True
print(f'Is "{string2}" lowercase?: {string2.islower()}') # string2 contains an uppercase letter, so it returns False
print(f'Is "{string3}" lowercase?: {string3.islower()}') # string3 contains lowercase letters and numbers, but it returns True because it only checks for lowercase letters and there is at least one lowercase letter in the string
print(f'Is "{string4}" lowercase?: {string4.islower()}') # string4 contains lowercase letters and a special character (@), but it returns True because it only checks for lowercase letters and there is at least one lowercase letter in the string
print(f'Is "{string5}" lowercase?: {string5.islower()}') # string5 is an empty string, so it returns False because there is no character in the string
print()
```

### isupper()

```python
# isupper() checks if all the characters in the string are uppercase letters (A-Z) and there is at least one character in the string

string1 = 'PYTHON'
string2 = 'Python'
string3 = 'PYTHON123'
string4 = 'PYTHON@123'
string5 = '' # empty string

print(f'Is "{string1}" uppercase?: {string1.isupper()}') # all characters in string1 are uppercase letters, so it returns True
print(f'Is "{string2}" uppercase?: {string2.isupper()}') # string2 contains a lowercase letter, so it returns False
print(f'Is "{string3}" uppercase?: {string3.isupper()}') # string3 contains uppercase letters and numbers, but it returns True because it only checks for uppercase letters and there is at least one uppercase letter in the string   
print(f'Is "{string4}" uppercase?: {string4.isupper()}') # string4 contains uppercase letters and a special character (@), but it returns True because it only checks for uppercase letters and there is at least one uppercase letter in the string
print(f'Is "{string5}" uppercase?: {string5.isupper()}') # string5 is an empty string, so it returns False because there is no character in the string
print()
```

### join()

```python
# join() joins the elements of an iterable (like list, tuple, etc.) into a single string, separated by the string on which it is called

string_list = ['hello', 'kemon', 'acho', 'tumi?']

print(string_list)
print(f'Joined string with space: {' '.join(string_list)}')
print(f'Joined string with hyphen: {'-'.join(string_list)}')
print(f'Joined string with underscore: {'_'.join(string_list)}')
print()
```

### strip()

```python
# strip() removes any leading (spaces at the beginning) and trailing (spaces at the end) characters (space is the default leading character to remove)
# lstrip() removes leading spaces only
# rstrip() removes trailing spaces only

string = '   Hello, World!   '
print(f'Original string: "{string}"')
print(f'String after strip(): "{string.strip()}"')
print()

# lstrip() removes leading spaces only
print(f'Original string: "{string}"')
print(f'String after lstrip(): "{string.lstrip()}"') 
print()

# rstrip() removes trailing spaces only
print(f'Original string: "{string}"')
print(f'String after rstrip(): "{string.rstrip()}"') 
print()

# strip() can remove given characters from the beginning and the end of the string

string = '---Hello, World!---'
print(f'Original string: "{string}"')
print(f'String after strip("-"): "{string.strip("-")}"')
print()
```
