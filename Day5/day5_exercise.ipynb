### 26. Join the following lists:
* front_end = ['HTML', 'CSS', 'JS', 'React', 'Redux']
* back_end = ['Node','Express', 'MongoDB']

### 27. After joining the lists in question 26. Copy the joined list and assign it to a variable full_stack, then insert Python and SQL after Redux.

```python
# 26.

print('26.')
front_end = ['HTML', 'CSS', 'JS', 'React', 'Redux']
back_end = ['Node', 'Express', 'MongoDB']
joined_list = front_end + back_end

print(f'Front end technologies: {front_end}')
print(f'Back end technologies: {back_end}')
print(f'Joined list of front end and back end technologies: {joined_list}')
print()

# 27.

print('27.')
print(f'Joined list before copying: {joined_list}')
full_stack = joined_list.copy() # copy() method creates a copy of the list
full_stack.append('Redux')
full_stack.append('Python')
full_stack.append('SQL')
print(f'Full stack technologies after appending: {full_stack}')
print()
```
