### 1. The following is a list of 10 students ages:

* Sort the list and find the min and max age

* Add the min age and the max age again to the list

* Find the median age (one middle item or two middle items divided by two)

* Find the average age (sum of all items divided by their number)

* Find the range of the ages (max minus min)

* Compare the value of (min - average) and (max - average), use abs() method

```python
ages = [19, 22, 19, 24, 20, 25, 26, 24, 25, 24]

print("Ages:", ages)
print()

# sorting the list and finding the min and max ages
ages.sort()
print("Sorted Ages:", ages)
print("Min Age:", ages[0])
print("Max Age:", ages[len(ages) - 1])
print()

# adding the min and max ages to the list
ages.append(ages[0])
ages.append(ages[len(ages) - 2])
print(f'Ages after adding min and max ages: {ages}')
print()

# finding the median age
ages.sort()
print("Sorted Ages:", ages)
print(f'Median Age: {ages[len(ages) // 2]}')
print()

# finding the average age
average_age = sum(ages) / len(ages)
print(f'Average Age: {average_age}')
print()

# finding the range of ages
print(f'Range of Ages: {ages[len(ages) - 1] - ages[0]}') # range is max age - min age
print()

# comparing the absolute differences between the min and average age, and the max and average age
min_age_diff = abs(ages[0] - average_age)
max_age_diff = abs(ages[len(ages) - 1] - average_age)
print(f'Absolute difference between min age and average age: {min_age_diff}')
print(f'Absolute difference between max age and average age: {max_age_diff}')
print()
```
