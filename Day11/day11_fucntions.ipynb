# Functions

```python
# functions

def add(x, y):
    return x + y

x = 5
y = 10

print(f'The sum of {x} and {y} using function call: {add(x, y)}\n')

# parameter less functions

def greet():
    print("Hello, welcome to Python programming!\n")

greet()

def question():
    return 'How are you doing?'

a = 'Shadipto'
print(f'{a}, {question()}\n')

def weight(mass, gravity):
    return mass * gravity

print(f'Weight of an obj with mass 22kg: {weight(22, 9.81)} N\n')
```

## Odd and Even fuction

```python
# odd or even function

def odd_or_even(num):
    if num % 2 == 0:
        return 'Even'
    else:
        return 'Odd'

print(f'The number is: {odd_or_even(89)}')
```

## Arbitary number of parameters

```python
# arbitary number of parametes

list1 = list()
def listup(*args):
    list1.append(args)

listup('hello', 'world', 'python')
print(f'List1: {list1}')

print()

def multiply(*args):
    result = 1
    for num in args:
        result *= num
    return result

product = multiply(2, 3, 4)
print(f'Multiplication of 2, 3 and 4 is: {product}\n')
```

## Dictionary unpacking

```python
# dictionary unpacking

my_dict = {'name': 'Alice', 'age': 30, 'city': 'New York'}

def print_info(name, age, city):
    print(f'Name: {name}')
    print(f'Age: {age}')
    print(f'City: {city}')

print_info(**my_dict)
```
