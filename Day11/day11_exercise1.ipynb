```python
# 1

print('1.')
def add_two_numbers(a, b):
    return a + b

result = add_two_numbers(3, 5)
print(f'The sum of 3 and 5 is: {result}')
print()

# 2.

import math
print('2.')
def area_of_circle(radius):
    return math.pi * radius ** 2

radius = 4
area = area_of_circle(radius)
print(f'The area of a circle with radius {radius} is: {area:.2f}')
print()

# 3.
print('3.')
def all_sum(*args):
    if all(type(arg) == int or type(arg) == float for arg in args): # check if all arguments are numbers
        return sum(args)
    else:
        return 'Please provide numbers only.'

result = all_sum(1, 2, 3, 4, 5)
error_result = all_sum(1, 'two', 3)
print(f'The sum of the numbers is: {result}')
print(f'Error result: {error_result}')
print()

# 4.
print('4.')
def celcius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

celsius = 25
print(f'{celsius}°C is equal to {celcius_to_fahrenheit(celsius)}°F\n')

# 5.
print('5.')
def season(month):
    if month in ['December', 'January', 'February']:
        return 'Winter'
    elif month in ['March', 'April', 'May']:
        return 'Spring'
    elif month in ['June', 'July', 'August']:
        return 'Summer'
    elif month in ['September', 'October', 'November']:
        return 'Autumn'
    else:
        return 'Invalid month'

month = 'April'
print(f'{month} is in {season(month)}\n')

# 6.
print('6.')
def slope(x1, y1, x2, y2):
    if x2 - x1 == 0:
        return 'Undefined (vertical line)'
    return (y2 - y1) / (x2 - x1)

x1, y1 = 1, 2
x2, y2 = 3, 4
print(f'The slope of the line connecting points ({x1}, {y1}) and ({x2}, {y2}) is: {slope(x1, y1, x2, y2)}\n')

# 7.
# ax^2 + bx + c = 0
# x1 = (-b + sqrt(b^2 - 4ac)) / 2a
# x2 = (-b - sqrt(b^2 - 4ac)) / 2a
print('7.')
def solve_quadratic(a, b, c):
    discriminant = pow(b, 2) - 4 * a * c
    if discriminant < 0:
        return 'No real roots'
    elif discriminant == 0:
        x = -b / (2 * a)
        return f'One real root: {x}'
    else:
        x1 = (-b + math.sqrt(discriminant)) / (2 * a)
        x2 = (-b - math.sqrt(discriminant)) / (2 * a)
        return f'Two real roots: {x1} and {x2}'

a, b, c = 1, -3, 2
print(f'The roots of the equation {a}x^2 + {b}x + {c} = 0 are: {solve_quadratic(a, b, c)}\n')

# 8.
print('8.')
list1 = ['apple', 'banana', 'cherry', 'date', 'fig', 'grape']
def print_list(list):
    for item in list:
        print(item)

print_list(list1)
print()

# 9.
print('9.')
def reverse_list(list):
    for i in range(len(list), 0, -1):
        print(list[i-1])

reverse_list(list1)
```

### 10. Declare a function named capitalize_list_items. It takes a list as a parameter and it returns a capitalized list of items

### 11. Declare a function named add_item. It takes a list and an item parameters. It returns a list with the item added at the end.

### 12. Declare a function named remove_item. It takes a list and an item parameters. It returns a list with the item removed from it.

```python
# 10.
print('10.')
list2 = ['apple', 'banana', 'cherry', 'date', 'fig', 'grape']
def capitalization(list):
    for item in list:
        print(item.capitalize())

capitalization(list2)
print()

# 11.
print('11.')
def add_item(list, item):
    list.append(item)
    return list

my_list = ['apple', 'banana', 'cherry']
print(f'Original list: {my_list}')
new_item = 'date'
updated_list = add_item(my_list, new_item)
print(f'Updated list: {updated_list}')
print()

# 12.
print('12.')
def remove_item(list, item):
    if item in list:
        list.remove(item)
        return list
    else:
        return f'Item "{item}" not found in the list.'

my_list = ['apple', 'banana', 'cherry', 'date']
print(f'Original list: {my_list}')
item_to_remove = 'banana'
updated_list = remove_item(my_list, item_to_remove)
print(f'Updated list after removing "{item_to_remove}": {updated_list}')
print()
```

### 13. Declare a function named sum_of_numbers. It takes a number parameter and it adds all the numbers in that range.

### 14. Declare a function named sum_of_odds. It takes a number parameter and it adds all the odd numbers in that range.

### 15. Declare a function named sum_of_even. It takes a number parameter and it adds all the even numbers in that range.

```python
# 13.
print('13.')
def sum_of_number(num):
    result = 0
    for i in range(0, num+1):
        result = result + i
    return result

a = sum_of_number(52)
print(f'The sum of numbers from 0 to 52 is: {a}')
print()

# 14.
print('14.')
def sum_of_odd_numbers(num):
    result = 0
    for i in range(1, num+1, 2):
        result = result + i
    return result

b = sum_of_odd_numbers(532)
print(f'The sum of odd numbers from 1 to 532 is: {b}\n')

# 15.
print('15.')
def sum_of_even_numbers(num):
    result = 0
    for i in range(0, num+1, 2):
        result = result + i
    return result

c = sum_of_even_numbers(532)
print(f'The sum of even numbers from 0 to 532 is: {c}\n')
```
