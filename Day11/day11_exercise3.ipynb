```python
# 1

print('1')
def isPrime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

print(f'1: {isPrime(1)}')
print(f'2: {isPrime(2)}')
print(f'3: {isPrime(3)}')
print(f'4: {isPrime(4)}')
print(f'5: {isPrime(5)}')
print(f'6: {isPrime(6)}')
print(f'7: {isPrime(7)}')

print()

# 2
print('2')
list1 = [1, 2, 3, 4, 5, 5, 6, 7, 8, 9, 10, 1, 2, 3, 4, 5, 5, 6, 7, 8, 9, 10]
list2 = [1, 2, 3, 4, 5]
def isUnique(lst):
    return len(lst) == len(set(lst))

print(f'list1: {list1}')
print(f'list1 isUnique: {isUnique(list1)}')
print(f'list2: {list2}')
print(f'list2 isUnique: {isUnique(list2)}\n')

# 3
print('3')
mixed_list = [1, 'hello', 3.14, True, None, [1, 2, 3], {'key': 'value'}, (1, 2), {1, 2, 3}]
list1 = [1, 2, 3, 4, 5]
def sametype(list):
    if not list:
        return 'list is empty'
    for item in range(0, len(list)):
        if type(list[item]) != type(list[0]):
            return False
    return True

print(f'mixed_list: {mixed_list}')
print(f'mixed_list sametype: {sametype(mixed_list)}\n')
print(f'list1: {list1}')
print(f'list1 sametype: {sametype(list1)}\n')

# 4
print('4')
def chk_variabel(variable):
    if variable.isidentifier():
        return True
    else:
        return False

print(f'chk_variabel("variable1"): {chk_variabel("variable1")}')
print(f'chk_variabel("1variable"): {chk_variabel("1variable")}')
print(f'chk_variabel("variable_1"): {chk_variabel("variable_1")}')
print(f'chk_variabel("variable-1"): {chk_variabel("variable-1")}')
print(f'chk_variabel("variable 1"): {chk_variabel("variable 1")}\n')
```
