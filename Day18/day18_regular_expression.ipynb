# Regular Expression

## Regular expressions cheat sheet:
### .       - Any character except newline
### ^       - Start of string
### $       - End of string
### *       - 0 or more repetitions
### +       - 1 or more repetitions
### ?       - 0 or 1 repetition
### {n}     - Exactly n repetitions
### {n,}    - n or more repetitions
### {n,m}   - Between n and m repetitions 
### []      - Character class
### |       - Alternation (OR)
### ()      - Grouping
### \       - Escape character
### \d      - Digit (0-9)
### \D      - Non-digit
### \w      - Word character (alphanumeric + underscore)
### \W      - Non-word character
### \s      - Whitespace character
### \S      - Non-whitespace character  

## re.match()

```python
# Regular expressions
# re.match() - matches a pattern at the beginning of the string
# re.search() - matches a pattern anywhere in the string
# re.findall() - finds all occurrences of a pattern in the string and returns them as a list

import re

txt = "The rain in Spain stays mainly in the plain"

pattern = r"The"
match = re.match(pattern, txt) # finds at the beginning of the string

print(f'Type of match: {type(match)}')
print(f'Match object: {match}')
print()

# Using re.search() to find a pattern anywhere in the string
pattern2 = r"rain"
match2 = re.search(pattern2, txt) # finds anywhere in the string
print(f'Type of match2: {type(match2)}')
print(f'Match2 object: {match2}')
print(f'Match2 span: {match2.span()}')
print()


# Using findall() to find all occurrences of a pattern in the string
txt2 = """
The rain in Spain stays mainly in the plain
The rain in Spain stays mainly in the plain
The rain in Spain stays mainly in the plain
"""

pattern3 = r"ain"
matches3 = re.findall(pattern3, txt2) # finds all occurrences of the pattern in the string
print(f'Type of matches3: {type(matches3)}')
print(f'Matches3 list: {matches3}')
print()

# using re.finditer() to find all occurrences of a pattern in the string and return an iterator
matches4 = re.finditer(pattern3, txt2) # finds all occurrences of the pattern in the string and returns an iterator
print(f'Type of matches4: {type(matches4)}')
print(f'Matches4 iterator: {matches4}')
for i in matches4:
    print(f'Match4 span: {i}') 
print()
```

## Replacing sub-string

```python
# replacing a pattern in the string using re.sub()
txt3 = "The rain in Spain stays mainly in the plain"
pattern4 = r"ain"
new_txt = "NEW"

new_txt3 = re.sub(pattern4, new_txt, txt3) # replaces all occurrences of the pattern in the string with the new string
print(new_txt3)
print()

# splitting a string using re.split()
txt4 = "The rain in Spain stays mainly in the plain"
pattern5 = r"\s" # split the string at every whitespace character
split_txt = re.split(pattern5, txt4) # splits the string at every whitespace character
print(f'Type of split_txt: {type(split_txt)}')
print(f'Split_txt list: {split_txt}')   
print()

# sub() with count parameter to limit the number of replacements

txt5 = '''%I a%m te%%a%%che%r% a%n%d %% I l%o%ve te%ach%ing.
T%he%re i%s n%o%th%ing as r%ewarding a%s e%duc%at%i%ng a%n%d e%m%p%ow%er%ing p%e%o%ple.
I fo%und te%a%ching m%ore i%n%t%er%%es%ting t%h%an any other %jobs.
D%o%es thi%s m%ot%iv%a%te %y%o%u to b%e a t%e%a%cher?'''

pattern6 = r'%'
new_txt5 = re.sub(pattern6, '', txt5) # replaces all occurrences of the pattern in the string with an empty string
print(new_txt5)
print()

new_txt6 = re.split('\n', new_txt5) # splits the string at every newline character
print(f'Type of new_txt6: {type(new_txt6)}')
print(f'New_txt6 list: {new_txt6}')
```
