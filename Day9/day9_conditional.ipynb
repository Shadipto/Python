# Conditional

```python
# conditional statements

# if statement

a = 3
print(f"a = {a}")
if a > 0:
    print("a is positive\n")

# if-else statement

b = -2
print(f"b = {b}")
if b > 0:
    print("b is positive\n")
else:
    print("b is negative\n")

# if-elif-else statement

c = 0
print(f"c = {c}")
if c > 0:
    print("c is positive\n")
elif c < 0:
    print("c is negative\n")
else:
    print("c is zero\n")

# nested if statement

d = 5
print(f"d = {d}")
if d > 0:
    if d % 2 == 0:
        print("d is a positive even number\n")
    else:
        print("d is a positive odd number\n")
elif d < 0:
    if d % 2 == 0:
        print("d is a negative even number\n")
    else:
        print("d is a negative odd number\n")
else:
    print("d is zero\n")


```

## AND / OR logic conditional

```python
# logical operators in conditional statements

user = 'admin'
password = 'admin123'
if user == 'adimn' and password == 'admin123':
    print("Access granted\n")
else:
    print("Access denied\n") 

age = 25
if age < 18 or age > 65:
    print("You are not eligible for this offer\n")
else:
    print("You are eligible for this offer\n")
```
