### 1. Write a code which gives grade to students according to theirs scores:

```python
# 1.
print('1.')

marks = float(input(f'Enter your marks: '))

if marks < 0 or marks > 100:
    print(f'Invalid marks\n')
elif marks >= 90:
    print(f'Grade: A\n')
elif marks >= 80:
    print(f'Grade: B\n')
elif marks >= 70:
    print(f'Grade: C\n')
elif marks >= 60:
    print(f'Grade: D\n')
else:
    print(f'Grade: F\n')
```

### 2. Get the month from user input then check if the season is Autumn, Winter, Spring or Summer. If the user input is: September, October or November, the season is Autumn. December, January or February, the season is Winter. March, April or May, the season is Spring June, July or August, the season is Summer

```python
# 2.
print('2.')
month = input(f'Enter the month: ').lower()

if month in ['september', 'october', 'november']:
    print(f'The season is Autumn\n')
elif month in ['december', 'january', 'february']:
    print(f'The season is Winter\n')
elif month in ['march', 'april', 'may']:
    print(f'The season is Spring\n')
elif month in ['june', 'july', 'august']:
    print(f'The season is Summer\n')
else:
    print(f'Invalid month\n')
```

### 3. fruits = ['banana', 'orange', 'mango', 'lemon']
### If a fruit doesn't exist in the list add the fruit to the list and print the modified list. If the fruit exists print('That fruit already exist in the list')

```python
# 3.
print('3.')

fruits = ['banana', 'orange', 'mango', 'lemon']

input_fruit = input(f'Enter a fruit: ').lower()

if input_fruit in fruits:
    print(f'{input_fruit} is in the list\n')
else:
    print(f'{input_fruit} is not in the list\n')
    print(f'Fruits list= {fruits}\n')
    fruits.append(input_fruit)
    print(f'Updated fruits list= {fruits}\n')
```
