### 1. Get user input using input(“Enter your age: ”). If user is 18 or older, give feedback: You are old enough to drive. If below 18 give feedback to wait for the missing amount of years.

```python
# 1.

print('1.')
age = int(input(f'Ente your age: '))

if age >= 18:
    print(f'You are old enough to drive\n')
else:
    print(f'You have to wait {18 - age} years to drive\n')
```

### 2. Compare the values of my_age and your_age using if … else. Who is older (me or you)? Use input(“Enter your age: ”) to get the age as input. You can use a nested condition to print 'year' for 1 year difference in age, 'years' for bigger differences, and a custom text if my_age = your_age.

```python
# 2.
print('2.')
my_age = 25
your_age = int(input(f'Enter your age: '))

if my_age > your_age:
    if my_age - your_age == 1:
        print(f'I am {my_age - your_age} year older than you\n')
    else:
        print(f'I am {my_age - your_age} years older than you\n')
elif my_age < your_age:
    if your_age - my_age == 1:
        print(f'You are {your_age - my_age} year older than me\n')
    else:
        print(f'You are {your_age - my_age} years older than me\n')
else:
    print(f'my age = your age\n')
```

### 3. Get two numbers from the user using input prompt. If a is greater than b return a is greater than b, if a is less b return a is smaller than b, else a is equal to b.

```python
# 3.
print('3.')
num1 = int(input(f'Enter first number: '))
num2 = int(input(f'Enter second number: '))

if num1 > num2:
    print(f'{num1} is greater than {num2}\n')
elif num1 < num2:
    print(f'{num2} is greater than {num1}\n')
else:
    print(f'{num1} is equal to {num2}\n')
```
