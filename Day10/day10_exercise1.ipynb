## 1. Iterate 0 to 10 using for loop, do the same using while loop.

## 2. Iterate 10 to 0 using for loop, do the same using while loop.

## 3. Write a loop that makes seven calls to print(), so we get on the output the following triangle:

### *
### **
### ***
### ****
### *****
### ******
### *******

```python
# 1.

print('1.')
print(f'Printing 0 to 10 using while loop')
i = 0
while True:
    if i > 10:
        break
    print(i)
    i += 1

print()

# 2.
print('2.')
print(f'Printing 0 to 10 using while loop')
i = i - 1
while True:
    print(i)
    i -= 1
    if i < 0:
        break
```

```python
# 3.
print('3.')

print(f'Printing pattern using for loop\n')
for i in range(1, 8):
    print(f'*' * i)
    print()
```

## 4. Use nested loops to create the following:
###  * * * * * * * *
###  * * * * * * * *
###  * * * * * * * *
###  * * * * * * * *
###  * * * * * * * *
###  * * * * * * * *
###  * * * * * * * *
###  * * * * * * * *

```python
# 4.

print('4.')
print(f'Printing pattern using for loop\n')
for i in range(1, 8):
    for j in range(1, 8):
        print(f'*', end=' ')
    print()
    print()
```

## 5. Print the following pattern:
### 0 x 0 = 0
### 1 x 1 = 1
### 2 x 2 = 4
### 3 x 3 = 9
### 4 x 4 = 16
### 5 x 5 = 25
### 6 x 6 = 36
### 7 x 7 = 49
### 8 x 8 = 64
### 9 x 9 = 81
### 10 x 10 = 100

```python
# 5.

print('5.')
print('Table pattern using for loop\n')
for i in range(1, 11):
    print(f'{i} x {i} = {i * i}\n')
    i += 1
    print()
```

### 6. Iterate through the list, ['Python', 'Numpy','Pandas','Django', 'Flask'] using a for loop and print out the items.

### 7. Use for loop to iterate from 0 to 100 and print only even numbers

### 8.Use for loop to iterate from 0 to 100 and print only odd numbers

```python
# 6.

list1 = ['Python', 'Numpy', 'Pandas', 'Django', 'Flask']
print('6.')
print(f'Printing list items using for loop\n')
for i in list1:
    print(i)

# 7.
print('7.')
print(f'Printing even numbers from 0 to 100 using for loop\n')
count = 1
for i in range(0, 101):
    if i % 2 == 0:
        print(f'Even number {count}:', i)
        count += 1
print()

# 8.
print('8.')
print(f'Printing odd numbers from 0 to 100 using for loop\n')
count = 1
for i in range(0, 101):
    if i % 2 != 0:
        print(f'Odd number {count}:', i)
        count += 1

print()
```
