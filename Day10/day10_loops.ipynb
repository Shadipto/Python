# Loops

## While loop

```python
# while loop

print(f'print numbers from 1 to 5 using while loop')
count = 1
while count <= 5:
    print(count)
    count += 1
print('End of while loop\n')

# while loop with else

print(f'print numbers from 1 to 5 using while loop with else')
count = 1
while count <= 5:
    print(count)
    count += 1
else:
    print('End of while loop with else\n')

# infinite while loop with break

print(f'print numbers from 1 to 5 using infinite while loop with break')
count = 1
while True:
    if count > 5:
        break
    print(count)
    count += 1
print('End of infinite while loop with break\n')
```

## While loop with condition

```python
# loops with continue

print(f'print odd numbers from 1 to 10 using while loop with continue')
count = 0
while count < 10:
    count += 1
    if count % 2 == 0:
        continue
    print(count)
print('End of while loop with continue\n')
```

## For loop

```python
# for loop

print(f'print numbers from 1 to 5 using for loop')
for i in range(1, 6):
    print(i)
print('End of for loop\n')

# for loop with else
print(f'print numbers from 1 to 5 using for loop with else')
for i in range(1, 6):
    print(i)
else:
    print('End of for loop with else\n')

# for loop with break
print(f'print numbers from 1 to 5 using for loop with break')
for i in range(1, 11):
    if i > 5:
        break
    print(i)
print('End of for loop with break\n')
```

## For loop for strings

```python
# for loop with string

string  = 'Hello, world. Welcome to Python programming.'

print(f'print each character of the string using for loop')
for char in string:
    print(char, end=' ')
print('\nEnd of for loop with string\n')

# alternative way to loop through a string

print(f'print each character of the string using alternative way to loop through a string')
for i in range(len(string)):
    print(string[i], end=' ')
print('\nEnd of alternative way to loop through a string\n')
```

## Loop with dictionay

```python
person = {
    'first_name':'Asabeneh',
    'last_name':'Yetayeh',
    'age':250,
    'country':'Finland',
    'is_marred':True,
    'skills':['JavaScript', 'React', 'Node', 'MongoDB', 'Python'],
    'address':{
        'street':'Space street',
        'zipcode':'02210'
    }
}
# for loop with dictionary

print(f'print each key of the dictionary using for loop')
for key in person:
    print(key)
print('End of for loop with dictionary\n')

print(f'print each value of the dictionary using for loop')
for val in person:
    print(person[val])
print('End of for loop with dictionary values\n')

# alternative way to loop through a dictionary

print(f'print each key of the dictionary using alternative way to loop through a dictionary')
for key in person.keys():
    print(key)
print('End of alternative way to loop through a dictionary keys\n')

print(f'print each value of the dictionary using alternative way to loop through a dictionary')
for val in person.values():
    print(val)
print('End of alternative way to loop through a dictionary values\n')
```

## Range function

```python
# range function in loops

print(f'print numbers from 0 to 4 using range function in for loop')
for i in range(5): # 0 to 4
    print(i)
print('End of for loop with range function\n')

print(f'print numbers from 2 to 6 using range function in for loop')
for i in range(2, 7): # 2 to 6
    print(i)
print('End of for loop with range function\n')

print(f'print numbers from 1 to 10 with a step of 2 using range function in for loop')
for i in range(1, 11, 2): # 1 to 10 with a step of 2
    print(i)
print('End of for loop with range function\n')
```
