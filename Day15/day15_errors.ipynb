# Types of Errors

```python
# missing parenthesis

print "Hello, World!" # This line is missing a opening and ending parenthesis
```

```python
# variable not defined

print(f'Age: {age}') # The variable 'age' is not defined before this line
```

```python
# index error

my_list = [1, 2, 3]
print(f'Printing the 4th element: {my_list[3]}') # This will raise an IndexError since the list has only 3 elements (index 0, 1, and 2)
```

```python
# module not found

import non_existent_module # This will raise a ModuleNotFoundError since there is no module named 'non_existent_module'

print("This line will not be executed due to the error above.")
```

```python
# attribute error

my_string = "Hello"
print(my_string.append(" World")) # This will raise an AttributeError since strings do not have an 'append' method
```

```python
# key error

my_dict = {"name": "Alice", "age": 30}
print(my_dict["gender"]) # This will raise a KeyError since there is no key named 'gender' in the dictionary
```

```python
# type error

result = 10 + "5" # This will raise a TypeError since you cannot add an integer and a string together
print(f'Result: {result}')
```

```python
# import error

from math import non_existent_function # This will raise an ImportError since there is no function named 'non_existent_function' in the math module
print("This line will not be executed due to the error above.")
```

```python
# value error

number = int("abc") # This will raise a ValueError since the string "abc" cannot be converted to an integer
print(f'Number: {number}')
```

```python
# zero division error

result = 10 / 0 # This will raise a ZeroDivisionError since division by zero is not allowed
print(f'Result: {result}')
```
