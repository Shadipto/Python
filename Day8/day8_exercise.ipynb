# Exercise

### 1. Create an empty dictionary called dog

### 2. Add name, color, breed, legs, age to the dog dictionary

### 3. Create a student dictionary and add first_name, last_name, gender, age, marital status, skills, country, city and address as keys for the dictionary

### 4. Get the length of the student dictionary

### 5. Get the value of skills and check the data type, it should be a list

### 6. Modify the skills values by adding one or two skills

### 7. Get the dictionary keys as a list

### 8. Get the dictionary values as a list

### 9. Change the dictionary to a list of tuples using items() method

### 10. Delete one of the items in the dictionary

### 11. Delete one of the dictionaries

```python
# 1.

print('1.')
dog = dict()
print(f'Empty dictionary: {dog}')
print(f'Data type of the dictionary: {type(dog)}')
print()

# 2.

print('2.')
dog['name'] = 'Buddy'
dog['age'] = 5
dog['breed'] = 'Golden Retriever'
print(f'Updated dictionary: {dog}')
print()

# 3.

print('3.')
student_dict = {
    'first_name': 'Shadipto',
    'last-name': 'Pranto',
    'gender': 'Male',
    'age': 23,
    'isMarried': False,
    'skills': ['C++', 'Python ', 'C#', 'JavaScript'],
    'country': 'Bangladesh',
    'city': 'Dhaka',
    'address': 'Kuril, Dhaka, Bangladesh'
}
print(f'Student dictionary: \n{student_dict}')
print()

# 4.
print('4.')
print(f'Length of the student dictionary: {len(student_dict)}')
print()

# 5.
print('5.')
print(f'Value of skills key: {student_dict["skills"]}')
print(f'Data type of skills key: {type(student_dict["skills"])}')
print()

# 6.
print('6.')
print('Adding new skills to the skills key...')
student_dict['skills'].extend(['HTML', 'CSS', 'React']) # .append() can also be used to add a single skill
print(f'Updated skills: {student_dict["skills"]}')
print()

# 7.
print('7.')
keys = list(student_dict.keys())
print(f'Keys of the student dictionary: {keys}')
print(f'Data type of keys: {type(keys)}')
print()

# 8.
print('8.')
values = list(student_dict.values())
print(f'Values of the student dictionary: {values}')
print(f'Data type of values: {type(values)}')
print()

# 9.
print('9.')
new_list = student_dict.items()
print(f'Items of the student dictionary: {new_list}')
print(f'Data type of items: {type(new_list)}')  
print()

# 10.
print('10.')
print('Deleting the student dictionary...')
del student_dict
try:
    print(student_dict)
except NameError:
    print('student_dict has been deleted and is no longer accessible.')
print()
```
