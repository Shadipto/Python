# Dictionaries

```python
# dictionaries are mutable, they can be changed after they are created
# contain key-value pairs, where each key is unique and maps to a value
# created using curly braces {} or the dict() constructor

# empty dictionary

empty_dict = {}
print(f'empty_dict: {empty_dict}') 
print()

empty_dict2 = dict()
print(f'empty_dict2: {empty_dict2}') 
print()

person = {
    'name': 'Shadipto',
    'age': 30,
    'city': 'Dhaka',
    'country': 'Bangladesh',
    'isStudent': False,
    'skills': ['Python', 'JavaScript', 'SQL', 'Java', 'C++'], # list in dictionary
    'education': {
        'degree': 'BsC in CSE',
        'university': 'AIUB',
        'graduation_year': 2027
    } # dictionary in a dictionary
}

print(f'person: {person}')
print()

print(f'Length of person dictionary: {len(person)}')
print()


```

## Accessing the elements dictionay 

```python
# indexig using keys

person = {
    'name': 'Shadipto',
    'age': 30,
    'city': 'Dhaka',
    'country': 'Bangladesh',
    'isStudent': False,
    'skills': ['Python', 'JavaScript', 'SQL', 'Java', 'C++'], # list in dictionary
    'education': {
        'degree': 'BsC in CSE',
        'university': 'AIUB',
        'graduation_year': 2027
    } # dictionary in a dictionary
}

# using keys to access values in the dictionary returns error if the key does not exist
print(f"person['name']: {person['name']}")
print(f"person['age']: {person['age']}")
print(f"person['city']: {person['city']}")
print(f"person['country']: {person['country']}")
print(f"person['isStudent']: {person['isStudent']}")
print(f"person['skills']: {person['skills']}")
print(f"person['education']: {person['education']}")
print()

# using get() method to access values in the dictionary returns None if the key does not exist
print(f"person.get('name'): {person.get('name')}")
print(f"person.get('age'): {person.get('age')}")
print(f"person.get('city'): {person.get('city')}")
print(f"person.get('country'): {person.get('country')}")
print(f"person.get('isStudent'): {person.get('isStudent')}")
print(f"person.get('skills'): {person.get('skills')}")
print(f"person.get('education'): {person.get('education')}")
print()
print(f'testing get() method with non-existing key: {person.get("non_existing_key")}')
print()

```

## Adding elements to the dictionay

```python
# adding new key-value pairs to the dictionary

person = {
    'name': 'Shadipto',
    'age': 30,
    'city': 'Dhaka',
    'country': 'Bangladesh',
    'isStudent': False,
    'skills': ['Python', 'JavaScript', 'SQL', 'Java', 'C++'], # list in dictionary
    'education': {
        'degree': 'BsC in CSE',
        'university': 'AIUB',
        'graduation_year': 2027
    } # dictionary in a dictionary
}

person['email'] = 'shadipto.pranto@gmail.com'
print(f'person after adding email: \n{person}')
print(f'person.get("email"): {person.get("email")}')
print()

# modifying existing key-value pairs in the dictionary

person['email'] = 'new@gmail.com'
print(f'person after modifying email: \n{person}')
print(f'person.get("email"): {person.get("email")}')
print()

# checking if a key exists in the dictionary

print(f'Is "name" in person? {"name" in person}')
print(f'Is "email" in person? {"email" in person}')
print(f'Is "zipcode" in person? {"zipcode" in person}')
print()

# removing key-value pairs from the dictionary

print('Removing "email" from person dictionary... using pop() method')
person.pop('email') # removes the key 'email' and its value from the dictionary
print(f'Is "email" in person after popping? {"email" in person}')
print()

# removing key-value pairs using popitem() method removes the last inserted key-value pair from the dictionary

print('Removing last inserted key-value pair from person dictionary... using popitem() method')
person.popitem() # removes the last inserted key-value pair from the dictionary
print(f'person after popping last item: \n{person}')
print()

# removing key-value pairs using del keyword

print('Removing "skills" from person dictionary... using del keyword')
del person['skills'] # removes the key 'skills' and its value from the dictionary
print(f'Is "skills" in person after deleting? {"skills" in person}')
print()
```

## Converting dictionaries to list

```python
# converting dictionary to list of tuples using items() method

person = {
    'name': 'Shadipto',
    'age': 30,
    'city': 'Dhaka',
    'country': 'Bangladesh',
    'isStudent': False,
    'skills': ['Python', 'JavaScript', 'SQL', 'Java', 'C++'], # list in dictionary
    'education': {
        'degree': 'BsC in CSE',
        'university': 'AIUB',
        'graduation_year': 2027
    } # dictionary in a dictionary
}

person_items = person.items() # returns a view object that displays a list of dictionary's key-value tuple pairs
print(f'person items: \n{person_items}')
print(f'data type of person_items: {type(person_items)}') # data type is dict_items
print()
print(f'data type of person_items: {type(person_items)}')
print(f'Length of person_items: {len(person_items)}')
print()

# copying dictionary using copy() method
person_copy = person.copy() # creates a shallow copy of the dictionary
print(f'person_copy: \n{person_copy}')
print()

# using clear() method to remove all items from the dictionary

print('Clearing all items from person dictionary... using clear() method')
person.clear() # removes all items from the dictionary
print(f'person after clearing: {person}')
print()
```

## Accessing keys and values

```python
person = {
    'name': 'Shadipto',
    'age': 30,
    'city': 'Dhaka',
    'country': 'Bangladesh',
    'isStudent': False,
    'skills': ['Python', 'JavaScript', 'SQL', 'Java', 'C++'], # list in dictionary
    'education': {
        'degree': 'BsC in CSE',
        'university': 'AIUB',
        'graduation_year': 2027
    } # dictionary in a dictionary
}

# accessing nested dictionary values
print(f"person['education']['degree']: {person['education']['degree']}")
print(f"person['education']['university']: {person['education']['university']}")
print(f"person['education']['graduation_year']: {person['education']['graduation_year']}")
print()

# accessing the keys as a list
person_keys = person.keys() # returns a view object that displays a list of all the keys
print(f'person keys: \n{person_keys}')
print(f'data type of person_keys: {type(person_keys)}')
print(f'Length of person_keys: {len(person_keys)}')
print()

# accessing the values as a list
person_values = person.values() # returns a view object that displays a list of all the values
print(f'person values: \n{person_values}')
print(f'data type of person_values: {type(person_values)}')
print(f'Length of person_values: {len(person_values)}')
print()
```
