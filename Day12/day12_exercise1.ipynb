### 1. Write a function which generates a six digit/character random_user_id.

```python
# 1

print('1.')
import random

# Generate a random user ID which is a six-digit number ordered from a set of characters (a-z, A-Z, 0-9)
def random_user_id():
    return ''.join(random.choices('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k = 6)) # Generate a random user ID of length 6

print(f'Random user ID: {random_user_id()}\n')

```

### 2. Modify the previous task. Declare a function named user_id_gen_by_user. It doesnâ€™t take any parameters but it takes two inputs using input(). One of the inputs is the number of characters and the second input is the number of IDs which are supposed to be generated.

```python
# 2

print('2.')
import random
def user_id_gen_by_user():
    char_num = int(input('Enter the number of characters for the user ID: '))
    num_of_ids = int(input('Enter the number of user IDs to generate: '))
    lsit = []
    for i in range(num_of_ids):
        lsit.append(''.join(random.choices('abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789', k = char_num)))
    return lsit

a = user_id_gen_by_user()
for i in a:
    print(f'Random user ID: {i}')
```

### 3. Write a function named rgb_color_gen. It will generate rgb colors (3 values ranging from 0 to 255 each).

```python
# 3
print('3.')
import random
def rgb_color_gen():
    r = random.randint(0, 255)
    g = random.randint(0, 255)
    b = random.randint(0, 255)
    return f'rgb({r}, {g}, {b})'

print(f'Random RGB color: {rgb_color_gen()}')
```
