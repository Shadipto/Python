# Exercise Level2

### 1. Write a function list_of_hexa_colors which returns any number of hexadecimal colors in an array (six hexadecimal numbers written after #. Hexadecimal numeral system is made out of 16 symbols, 0-9 and first 6 letters of the alphabet, a-f. Check the task 6 for output examples).

### 2. Write a function list_of_rgb_colors which returns any number of RGB colors in an array.

### 3. Write a function generate_colors which can generate any number of hexa or rgb colors.

```python
# 1
print('1.')
import random
def list_of_hexa(n):
    hex_list = list()
    for i in range(n):
        hex_list.append(f'#{"".join(random.choices("0123456789abcdef", k = 6))}')
    return hex_list

a = list_of_hexa(5)
print('List of 5 random hexadecimal colors:')
for i in a:
    print(i)
print()

# 2
print('2.')
def list_of_rgb(n):
    rgb_list = list()
    for i in range(n):
        r = random.randint(0, 255)
        g = random.randint(0, 255)
        b = random.randint(0, 255)
        rgb_list.append(f'rgb({r}, {g}, {b})')
    return rgb_list

b = list_of_rgb(5)
print('List of 5 random RGB colors:')
for i in b:
    print(i)
print()

# 3
print('3.')
def generate_colors(type, n):
    if type == 'hexa':
        return list_of_hexa(n)
    elif type == 'rgb':
        return list_of_rgb(n)
    else:
        return 'Invalid color type. Please choose "hexa" or "rgb".'

print('List of 5 random hexadecimal colors:')
for i in generate_colors('hexa', 5):
    print(i)
print()

print('List of 5 random RGB colors:')
for i in generate_colors('rgb', 5):
    print(i)
print()
```
