# Print syntax [uncommon]

```python
print('hello') # Normal print statement

print('hello', 'world', "!") # Print two strings separated by a space using comma [ , ]

print(f'hello {"world"} !') # Print two strings using f-string formatting

print(f"hahaha {2}") # Print string with an integer using f-string formatting
```

# Casting

```python
# int to float

a = 48
print(f'Int a: {a}')
print(f'Float a: {float(a)}')
print()

# float to int

b = 3.1416
print(f'Float b: {b}')
print(f'Int b: {int(b)}')
print()

# int to string

c = 12423
print(f'Int c: {c}')
print(f'String c: {str(c)}')

# string to list

d = 'abcdehahahaha'
print(f'String d: {d}')
print(f'List d: {list(d)}')
print()

# list to string
e = ['a', 'b', 'c', 'd', 'e']
print(f'List e: {e}')
# "" is the separator and join() is the method to convert list to string each element is concatenated what is inside the "" as separator
print(f'String e: {''.join(e)}') 
print(f'String e: {"-".join(e)}') # using - as separator and '' and "" both works
print()
```
