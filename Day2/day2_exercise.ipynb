```python
firstName = 'Shadipto'
lastName = "Pranto"
fullName = firstName + ' ' + lastName # used concatenation to join two strings
country = 'Bangladesh'
city = 'Dhaka'
age = 22
year = 2026
is_married = False
is_true = True
is_light_on = True

# 1. checking data types

print('1.')
print(f'Type of firstName: {type(firstName)}')
print(f'Type of lastName: {type(lastName)}')
print(f'Type of fullName: {type(fullName)}')
print(f'Type of country: {type(country)}')
print(f'Type of city: {type(city)}')
print(f'Type of age: {type(age)}')
print(f'Type of year: {type(year)}')
print(f'Type of is_married: {type(is_married)}')
print(f'Type of is_true: {type(is_true)}')
print(f'Type of is_light_on: {type(is_light_on)}')
print()

# 2. length of firstName

print('2.')
print(f'Length of firstName: {len(firstName)}')
print()

# 3. compair length of firstName and lastName

print('3.')
print(f'Length of firstName: {len(firstName)} & Length of lastName: {len(lastName)}')
if len(firstName) > len(lastName):
    print(f'First name is longer than last name.')
elif len(firstName) < len(lastName):
    print(f'Last name is longer than first name.')
else:
    print(f'First name and last name are of equal length.')
print()

# 4. 

num_one = 5
num_two = 4

# 5. sum num_one and num_two

print('5.')
total = num_one + num_two
print(f'The sum of num_one and num_two is: {total}')
print()

# 6. substract num_two from num_one

print('6.')
diff = num_two - num_one
print(f'The difference when num_two is subtracted from num_one is: {diff}')
print()

# 7. product of num_one and num_two

print('7.')
product = num_one * num_two
print(f'The product of num_one and num_two is: {product}')
print()

# 8. division of num_one by num_two

print('8.')
divion = num_one / num_two
print(f'The division of num_one by num_two is: {divion}')
print()

# 9. remainder of num_one divided by num_two

print('9.')
remainder = num_one % num_two
print(f'The remainder of num_one divided by num_two is: {remainder}')
print()

# 10. num_one to the power of num_two

print('10.')
variable_exp = num_one ** num_two
print(f'num_one to the power of num_two is: {variable_exp}')
print()

# 11. floor division of num_one by num_two

print('11.')
floor_division = num_one // num_two
print(f'The floor division of num_one by num_two is: {floor_division}')
print()

# 12. circle radius is 30 meters

print('12. (i)')

pi = 3.1416
radius = 30
area_of_circle = (pi*radius)**2
print(f'The area of the circle with radius {radius} meters is: {area_of_circle} square meters.')
print()

print('12. (ii)')
circum = 2*pi*radius
print(f'The circumference of the circle with radius {radius} meters is: {circum} meters.')
print()

print('12. (iii)')
radiusInput = float(input('Enter the radius of the circle in meters: '))
area_of_circle_input = (pi*(radiusInput))**2
print(f'The area of the circle with radius {radiusInput} meters is: {area_of_circle_input} square meters.')
print()

# 13. using input() function

print('13.')
input_firstName = input('Enter your first name: ')
input_lastName = input('Enter your last name: ')
input_country = input('Enter your country: ')
input_city = input('Enter your city: ')
input_age = int(input('Enter your age: ')) # input() function returns a string so we need to convert it to int using int() function
input_year = int(input('Enter the year you were born: '))

print(f'First Name: {input_firstName}')
print(f'Last Name: {input_lastName}')
print(f'Country: {input_country}')
print(f'City: {input_city}')
print(f'Age: {input_age}')
print(f'Year: {input_year}')
print()
    
```
